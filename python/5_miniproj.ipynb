```python
playstore = playstore.rename(columns= str.lower).drop_duplicates(subset='app')
```

```python
playstore
```

```python
playstore[['App', 'Type']].drop_duplicates(
).value_counts('Type', normalize=True)
```

```python
playstore['price'] = playstore.price.str.strip('$').apply(float)
```

```python
playstore[(playstore['Category'] == 'EDUCATION') & (playstore['Reviews'] > 1000)].reset_index()
```

```python
result = (
    playstore
    .groupby(['category', 'type'])
    .agg(mean_price=('price', 'mean'), mean_rating=('rating', 'mean'), mean_reviews=('reviews', 'mean'))
)
```

```python
result = result.round(2)
result['mean_rating'] = result.mean_rating.round(1)
```

```python
result
```

```python
result.to_csv('5_miniproj.csv', sep=',', index=True)
```

```python
result_2 = pd.pivot_table(
    playstore, 
    values=['price', 'rating', 'reviews'], 
    index=['category', 'type'], 
    aggfunc={'price': lambda x: round(np.mean(x), 2),
             'rating': lambda x: round(pd.mean(x), 1),
             'reviews': lambda x: round(pd.mean(x), 2)}
)

result_2 = result_2.rename(
    columns={'price': 'mean_price', 'rating': 'mean_rating', 'reviews': 'mean_reviews'})
```

```python
result_2.to_csv('5_miniproj2.csv', sep=',')
```
